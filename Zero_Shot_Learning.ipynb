```python
from transformers import pipeline

classifier = pipeline ("zero-shot-classification","facebook/bart-large-mnli")
```

```python
#For Single Labels
sentence="I have seen the cats and dogs"
candidate_labels=["sports","plants","education","animals"]

classifier(sentence,candidate_labels)
```

```python
# For Multilables
sentence="One day I will see the world"
candidate_labels=["travel","cooking","dansing","exploration","globalization"]

classifier(sentence,candidate_labels,multi_label=True)
```
